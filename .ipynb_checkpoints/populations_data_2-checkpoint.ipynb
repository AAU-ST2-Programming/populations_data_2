# Øvelse efter forelæsningen (scikit‑learn)

## Martins ECG‑Beat Bonanza  
**Scenarie:**  
Martin kommer med to NumPy‑arrays: peak amplitude og beat width for tusindvis af hjerteslag.  
Han siger:  
“Jeg vil gerne vide, om der er naturlige grupper af hjerteslag. Brug k‑means i Python.”

**Datastruktur:**  
```python
peak_amplitude = np.array([...])
beat_width = np.array([...])
```

**Opgave:**  
- Kombinér arrays til en 2D NumPy‑matrix.  
- Kør k‑means med forskellige værdier af k.  
- Visualisér klynger.  
- Forklar, hvad grupperne kunne repræsentere.

**Eksempel på løsning:**  
```python
import numpy as np
from sklearn.cluster import KMeans

X = np.column_stack((peak_amplitude, beat_width))

model = KMeans(n_clusters=3, random_state=0)
labels = model.fit_predict(X)

centroids = model.cluster_centers_
```

---

# Evaluering af k‑means

## Inertia  
Summen af kvadrerede afstande fra hvert punkt til dets centroid.  
Bruges til elbow‑metoden.

## Silhouette score  
Et mål for hvor godt et punkt passer ind i sin klynge sammenlignet med andre klynger.  
Værdier mellem −1 og 1.

## Evaluering når man har labels  
Når man har “ground truth” labels, kan man sammenligne direkte.

- Accuracy  
- Confusion matrix  
- Rand index / Adjusted Rand index  

**Eksempel:**  
```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy = accuracy_score(true_labels, labels)
cm = confusion_matrix(true_labels, labels)
```
